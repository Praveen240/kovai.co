```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from datetime import timedelta
```

```python
# Load the dataset
data = pd.read_csv(r'C:\Users\USER\OneDrive\Desktop\Praveen_V-21AD036\kovai.co_dataset.csv')

# Convert the 'Date' column to datetime and set it as the index
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
data.set_index('Date', inplace=True)

# Display basic statistics and the first few rows
print("Dataset Summary:")
print(data.describe())
print("\nFirst few rows of the dataset:")
print(data.head())

# ----------------------
# Task 1: Insights
# ----------------------

# 1. Plot trends in each service type
plt.figure(figsize=(15, 10))
for i, column in enumerate(data.columns[:-1], 1):  # Exclude 'Other' column
    plt.subplot(3, 2, i)
    plt.plot(data.index, data[column], label=column, color='b')
    plt.title(f'Trend in {column}')
    plt.xlabel('Date')
    plt.ylabel('Count')
    plt.legend()
plt.tight_layout()
plt.show()

# 2. Key statistical insights
print("\nAverage ridership by service type:")
print(data.mean())

print("\nPeak ridership day for each service type:")
print(data.idxmax())

print("\nCorrelation between different service types:")
print(data.corr())

# Visualize correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between service types")
plt.show()


# 3. Box Plots for each service type
plt.figure(figsize=(12, 6))
sns.boxplot(data=data)
plt.title('Box Plot of Ridership by Service Type')
plt.ylabel('Ridership Count')
plt.xlabel('Service Type')
plt.show()
```

```python
# ----------------------
# Task 2: Forecasting
# ----------------------

# Define a function to forecast each service category for the next 7 days using ARIMA
def forecast_series(data, column_name, steps=7):
    # Fit ARIMA model
    model = ARIMA(data[column_name], order=(5,1,1))
    model_fit = model.fit()

    # Forecast the next 'steps' days
    forecast = model_fit.forecast(steps=steps)
    forecast_dates = [data.index[-1] + timedelta(days=i) for i in range(1, steps + 1)]
    
    # Plot the forecast
    plt.figure(figsize=(10, 5))
    plt.plot(data[column_name], label='Historical')
    plt.plot(forecast_dates, forecast, label='Forecast', color='red')
    plt.title(f'7-Day Forecast for {column_name}')
    plt.xlabel('Date')
    plt.ylabel('Count')
    plt.legend()
    plt.show()
    
    return forecast

# Generate and display forecasts for each required service type
forecasts = {}
forecast_df = pd.DataFrame(index=[data.index[-1] + timedelta(days=i) for i in range(1, 8)])

for column in ['Local Route', 'Light Rail', 'Peak Service', 'Rapid Route', 'School']:
    print(f"\nForecast for {column}:")
    forecasts[column] = forecast_series(data, column)
    forecast_df[column] = forecasts[column]

print("\n7-Day Forecast Data:")
print(forecast_df)
```
